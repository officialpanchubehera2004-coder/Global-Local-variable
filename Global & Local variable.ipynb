# Global Keywords

#The global keyword is used inside a function to indicate that a variable refers to a global varibale not a local variable.

```python
x = 10  # Global variable

def update_x():
    #global x  # Declare that we are using the global variable x
    x += 5    # Modify the global variable

update_x()
print(x)  # Output: 15
```

```python
x = 10  # Global variable

def update_x():
    global x  # Declare that we are using the global variable x
    x += 5    # Modify the global variable

update_x()
print(x)  # Output: 15
```

```python
x = 10  # Global variable

def update_x():
    globals()['x'] += 5  # Access and modify the global variable using the dictionary

update_x()
print(x)  # Output: 15
```

# Pass list to function

#can we pass list of element in the function that function will return the count of even or odd number from the list.

```python
def count(lst):
    
    even = 0
    odd = 0
    
    for i in lst:
        if i%2 == 0:
            even += 1
        else:
            odd +=1 
    return even,odd

lst = [10, 9, 8, 23, 50, 4, 8, 7 ]
even, odd = count(lst)         

print(even)
print(odd)
```

# Fibonacci Sequence

```python
# in progammin we need to continue these process thats why we need to use loop hear

def fib(n):
    a = 0
    b = 1
    
    print(a)
    print(b)
    
    for i in range(0, n):
        c = a + b
        a = b
        b = c
        
        print(c)
        
fib(10)
```

# Factorial of a number in python

```python
def fact(n):
    f=1
    for i in range(1,n+1):
        f=f*i
    return f

x =4
result = fact(x)
print(result)
```
